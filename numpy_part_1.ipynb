##TensorFlow Fundamentals - Scalars, Vectors, Matrices, and Tensors


```python
# Import TensorFlow
import tensorflow as tf
print(tf.__version__)
```

### Scalar/Constant

```python
# Create a scalar
scalar = tf.constant(24)
scalar
```

```python
# Check the number of dimensions (rank) of the scalar
scalar.ndim  # Also the Rank of the scalar
```

##Vector

```python

# Create a vector
vector = tf.constant([12, 24])
vector
```

```python
# Check the number of dimensions (rank) of the vector
vector.ndim  # Rank
```


## Matrix


```python
matrix = tf.constant([[12, 24],
                      [36, 48]])
matrix
```

```python
matrix.ndim  # Rank
```

```python
# Create another matrix with 3 rows and 2 columns
one_more_matrix = tf.constant([[1,2,3],
                               [4,5,6]])
one_more_matrix
```

```python
# Check the number of dimensions
one_more_matrix.ndim
```

## # Tensor

```python
# Create a tensor (3D or higher dimensional array)
tensor = tf.constant([[[1,2,3],
                       [4,5,6]],

                      [[4,5,6],
                       [7,8,9]],

                      [[7,8,9],
                       [10,11,12]]])
tensor
```

```python
# Check the number of dimensions (rank) of the tensor
tensor.ndim
```

```python
print("\n" + "="*50)
print("SUMMARY")
print("="*50)
print(f"Scalar: {scalar.numpy()}, ndim: {scalar.ndim}, shape: {scalar.shape}")
print(f"Vector: {vector.numpy()}, ndim: {vector.ndim}, shape: {vector.shape}")
print(f"Matrix: \n{matrix.numpy()}, \nndim: {matrix.ndim}, shape: {matrix.shape}")
print(f"Tensor shape: {tensor.shape}, ndim: {tensor.ndim}")
print("="*50)
```
